```python
print("\nPart 1:\n")

#Variable assignment
temperatures_C = [33,66,65,0,59,60,62,64,70,76,80,69,80,83,68,79,61,53,50,49,53,48,45,39]

missing_data_point_list = []
temperatures_F = []
dangerous_temperatures = []
count = []

#Initial Calculations

def temperatures_greater_70():
  for i in temperatures_C:
    if i > 70:
      print(str(i) + "ºC")
    elif i < 70:
      continue

#Setting high temperatures list
for i in temperatures_C:
  if i >= 70:
    dangerous_temperatures.append(i)
  elif i < 70:
    continue

#Setting missing data point list to correct list error
for i in temperatures_C:
  if len(missing_data_point_list) < 3:
    missing_data_point_list.append(i)
  elif len(missing_data_point_list) > 3:
    break

missing_data_point =  (sum(missing_data_point_list))/(len(missing_data_point_list))

temperatures_C[temperatures_C.index(0)] = round(missing_data_point)

temperatures_min = min(temperatures_C)
temperatures_max = max(temperatures_C)
temperatures_mean = (sum(temperatures_C))/(len(temperatures_C))

#Converting ºC to ºF
for i in temperatures_C:
  f = 1.8 * i + 32
  temperatures_F.append(round(f, 2))

#Printing calculation results
print("\nMin temperature: " + str(temperatures_min) + "ºC")
print("\nMax temperature: " + str(temperatures_max) + "ºC")
print("\nAverage temperature: " + str(round(temperatures_mean, 2)) + "ºC")
print("\nTemperatures greater than 70ºC that day:")
temperatures_greater_70()

print("\nProcessor temperatures in ºC:", temperatures_C)
print("\nProcessor temperatures in ºF:", temperatures_F)

#Checking conditions to evaluate cooling system 
if len(dangerous_temperatures) > 4 or max(dangerous_temperatures) > 80 or (sum(temperatures_C)/len(temperatures_C)) > 65:
  print("\nTrue, the cooling system should be changed")
else:
  print("\nFalse, the cooling system should not be changed")

#----------------------------------------------------------------------------------------------------------------------

print("\nPart 2:\n")

import statistics 

#We want the hours (not the temperatures) whose temperature exceeds 70ºC
for i in temperatures_C:
  count.append(i)
  if i > 70:
    print("At", (len(count) - 1), "hours the temperature = " + str(i) + "ºC")
  else:
    continue

#Condition that those hours are more than 4 consecutive and consecutive, not simply the sum of the whole set. Is this condition met?

#Average of each of the lists (ºC and ºF). How they relate?
print("\nAverage temperature in ºC:", round(statistics.mean(temperatures_C)))
print("\nAverage temperature in ºF:", round(statistics.mean(temperatures_F)))

#Standard deviation of each of the lists. How they relate?
print("\nTemperature's standard deviation in ºC:", round(statistics.stdev(temperatures_C)))
print("\nTemperature's standard deviation in ºF:", round(statistics.stdev(temperatures_F)))
```
